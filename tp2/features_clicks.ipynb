```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as mplt
import matplotlib as plt
import datetime as dati
sns.set()
```

```python
clicks = pd.read_csv("data/clicks.csv", dtype={'advertiser_id': 'category', 'action_id': 'float64', 'source_id': 'category',
                                               'created': 'str', 'country_code':'category', 'latitude': 'float64',
                                               'longitude':'float64', 'wifi_connection':'bool' , 'carrier_id':np.float16,
                                               'trans_id':'str', 'os_minor':'float64', 'agent_device':'float64', 'os_major':'float64',
                                               'specs_brand':'int64', 'brand':'category', 'timeToClick':'float64',
                                                'ref_type':'category', 'ref_hash':'int64'}, parse_dates = ['created'])
clicks.dtypes
```

```python
clicks['agent_device'].value_counts()
```

```python
clicks=pd.get_dummies(clicks,columns=['wifi_connection','specs_brand'])
```

```python
clicks.head()
```

```python
clicks['mean_hour']=(clicks['created'].dt.hour+clicks['created'].dt.minute/60+clicks['created'].dt.second/3600).astype(float)
clicks['std_hour']=(clicks['created'].dt.hour+clicks['created'].dt.minute/60+clicks['created'].dt.second/3600).astype(float)
```

```python
clicks=clicks.sort_values(by='created')
```

```python
mean_hour_gb=clicks.groupby(['ref_hash'])['mean_hour'].expanding(2).mean()
```

```python
std_hour_gb=clicks.groupby(['ref_hash'])['std_hour'].expanding(2).std()
```

```python
clicks_time_to_click_mean=clicks.groupby(['ref_hash'])['timeToClick'].expanding(2).mean()
```

```python
clicks_time_to_click_std=clicks.groupby(['ref_hash'])['timeToClick'].expanding(2).mean()
```

```python
clicks['time_to_click_min']=clicks['timeToClick']
```

```python
clicks['time_to_click_max']=clicks['timeToClick']
```

```python
features_time_to_click=clicks.groupby('ref_hash').agg({'time_to_click_min':'cummin', 'time_to_click_max':'cummax'})
```
