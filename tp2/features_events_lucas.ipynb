```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math as mt
%matplotlib inline
```

```python
events = pd.read_csv("data/events.csv.gzip",compression='gzip', dtype={'date': 'str', 'event_id': np.int16, 'ref_type': 'category',
                                               'ref_hash': 'category', 'application_id':np.int16, 'attributed': 'bool',
                                               'device_countrycode': 'int64', 'device_os_version':'float64', 'device_brand': 'float64',
                                               'device_model': 'float64', 'device_city':'float64', 'session_user_agent': 'float64',
                                               'trans_id': 'str', 'user_agent':'float64', 'event_uuid': 'str',
                                               'carrier': 'float64', 'kind':'float64', 'device_os': 'category',
                                                'connection_type':'str', 'ip_address': 'int64',
                                               'device_language': 'float64'}, parse_dates = ['date'],low_memory=False)
events.drop('device_os',axis=1,inplace=True)
events.drop('session_user_agent',axis=1,inplace=True)
events.drop('user_agent',axis=1,inplace=True)
events.drop('kind',axis=1,inplace=True)
events.drop('device_countrycode',axis=1,inplace=True)
events.drop('device_os_version',axis=1,inplace=True)

events.head(5)
```

```python
events=pd.get_dummies(events, columns=[ "attributed"])
```

```python
events['contador']=1
```

```python
events_attribuidos=events[events['attributed_True']==1]
```

```python
lista_id_eventos_atribuidos=events_attribuidos['event_id'].value_counts().index.to_list()
```

```python
events['eventos_utiles']=events['event_id'].isin(lista_id_eventos_atribuidos)
```

```python
events=events.sort_values(by='date')
events=events.reset_index()
events.head()
```

```python
events=events.drop(['level_0','index'],axis=1)
events.head()
```

```python
auctions=pd.get_dummies(events, columns=[ "eventos_utiles"])
```

```python
contador_ip=events[['ref_hash','ip_address']].drop_duplicates().groupby('ref_hash').cumcount() + 1
contador_ip=contador_ip.reindex(events.index).ffill()
```

```python
contador_ip.head()
```

```python
contador_ip=contador_ip.reset_index()
contador_ip=contador_ip.rename(columns = {0:'ip_distintas_ingresadas'})
```

```python
contador_ip=contador_ip.set_index('index')
contador_ip.head()
```

```python
contador_ip.to_csv('data/ip_distintas.csv')
```

```python
contador_ciudades=events[['ref_hash','device_city']].drop_duplicates().groupby('ref_hash').cumcount() + 1
contador_ciudades=contador_ciudades.reindex(events.index).ffill()
```

```python
contador_ciudades=contador_ciudades.to_frame()
```

```python
contador_ciudades=contador_ciudades.rename(columns = {0:'ciudades_distintas_ingresadas'})
```

```python
contador_aplicaciones=events[['ref_hash','application_id']].drop_duplicates().groupby('ref_hash').cumcount() + 1
contador_aplicaciones=contador_aplicaciones.reindex(events.index).ffill()
```

```python
contador_aplicaciones=contador_aplicaciones.to_frame()
contador_aplicaciones.head()
```
