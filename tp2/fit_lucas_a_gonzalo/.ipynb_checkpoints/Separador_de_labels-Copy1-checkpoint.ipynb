```python
import pandas as pd
import numpy as np
```

```python
targets = pd.read_csv('../target_competencia_ids.csv', dtype={'ref_hash': np.object, 'obj': np.int8})
targets.head()
```

```python
targets['tipo'] = targets['ref_hash'].str.split('_').str[1]
targets.head()
```

```python
targets['ref_hash'] = targets['ref_hash'].str.split('_').str[0]
targets.head()
```

```python
target_st = targets.loc[targets['tipo'] == 'st']
target_st = target_st.drop('tipo', axis=1).set_index('ref_hash')
target_st.to_csv('../xgb/base_target_st.csv')
target_st.head()
```

```python
target_sc = targets.loc[targets['tipo'] == 'sc']
target_sc = target_sc.drop('tipo', axis=1).set_index('ref_hash')
target_sc.to_csv('../xgb/base_target_sc.csv')
target_sc.head()
```
